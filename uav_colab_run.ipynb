# AVSCA — UAV Dataset Fusion Pipeline

Run the three cells below **sequentially** on a GPU-enabled Colab runtime.

| Cell | Action |
|------|--------|
| 1 | Mount Google Drive |
| 2 | Clone repo & install dependencies |
| 3 | Run fusion pipeline → zip output → save to Drive |

```python
# Cell 1 — Mount Google Drive
# Your dataset zips must already be in My Drive/UAV_Data/Raw/
from google.colab import drive
drive.mount('/content/drive')
```

```python
%%bash
# Cell 2 — Clone repository and install dependencies
git clone https://github.com/pratoshhhh/uav-dataset-fusion /content/uav-dataset-fusion
cd /content/uav-dataset-fusion
pip install -r requirements.txt
```

```python
%%bash
# Cell 3 — Run fusion pipeline, then zip and back up to Google Drive
python /content/uav-dataset-fusion/src/data_fusion.py \
  --visdrone_dir /content/drive/MyDrive/UAV_Data/Raw/VisDrone.zip \
  --heridal_dir  /content/drive/MyDrive/UAV_Data/Raw/Heridal.zip \
  --ttpla_dir    /content/drive/MyDrive/UAV_Data/Raw/TTPLA.zip \
  --wisard_dir   /content/drive/MyDrive/UAV_Data/Raw/WiSARD.zip \
  --output_dir   /content/master_uav_dataset

zip -r /content/master_uav_dataset.zip /content/master_uav_dataset

cp /content/master_uav_dataset.zip \
   /content/drive/MyDrive/UAV_Data/Ready/master_uav_dataset.zip

echo "Done. Dataset saved to My Drive/UAV_Data/Ready/master_uav_dataset.zip"
```
